<p style="font-size:28px; font-weight:bold; color:red"><i>6. Aggregate</i></p>

```python
departement.aggregate([
    {
        "$unwind": "$employes"  
    },
    {
        "$group": {
            "_id": "$chef_departement",
            "total_salaire": {"$sum": "$employes.salaire"}
        }
    }
])
for document in departement.aggregate([
    {
        "$unwind": "$employes"  
    },
    {
        "$group": {
            "_id": "$chef_departement",
            "total_salaire": {"$sum": "$employes.salaire"}
        }
    }
]):
    print(document)
```

<ul><li style="color: green; font-size:25px;">Aggregation avec Graphe</li></ul>

```python
import matplotlib.pyplot as plt

def custom_aggregation(departement):
    pipeline = [
        {
            "$unwind": "$employes"
        },
        {
            "$group": {
                "_id": "$chef_departement",
                "total_salaire": {"$sum": "$employes.salaire"}
            }
        }
    ]
    result = departement.aggregate(pipeline)
    return result

# Call your custom aggregation function
result = custom_aggregation(departement)

# Extract data for plotting
l_chef_departement = []
l_total_salaire = []
for document in result:
    l_chef_departement.append(document["_id"])
    l_total_salaire.append(document["total_salaire"])

# Create the graph
plt.figure(figsize=(10, 5))
plt.bar(l_chef_departement, l_total_salaire, color='green', width=0.4)
plt.xlabel("Chef Departement")
plt.ylabel("Total Salaire")
plt.title("Total Salaire for Each Chef Departement")
plt.show()
```
